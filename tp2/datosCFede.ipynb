```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfAvisosDet = pd.read_csv("procesado1/pAvisos1.csv")
dfPostu = pd.read_csv("todos/PPostu.csv")
```

```python
dfPostuXAviso = dfPostu.loc[:,["idaviso","idpostulante"]].groupby("idaviso").count().rename(columns = {"idpostulante":"cantidad"}).reset_index()
```

```python
dfmerge = pd.merge(dfPostuXAviso,dfAvisosDet,on = "idaviso",how="right")
dfmerge["cantidad"] = dfmerge["cantidad"].fillna(0)
```

```python
dfV = pd.read_csv("todos/Vistas.csv")
#dfA = pd.read_csv("todos/AvisosDet.csv")
```

```python
dfPostu["fechapostulacion"] = pd.to_datetime(dfPostu["fechapostulacion"])
```

```python
dfV["timestamp"] = pd.to_datetime(dfV["timestamp"])
```

```python
dfV.to_csv("procesado1/VistasCCorrecta.csv")
```

```python
dfPostuMin = dfPostu.loc[:, ['idaviso', 'fechapostulacion']].groupby('idaviso').min().reset_index()
```

```python
dfVMin = dfV.rename(columns={'timestamp':'fechapostulacion', 'idAviso':'idaviso'})\
                        .loc[:, ['idaviso', 'fechapostulacion']].groupby('idaviso').min().reset_index()
```

```python
dfPostuMin.shape
```

```python
dfVMin.shape
```

```python
fechas_iniciales = pd.concat([dfPostuMin, dfVMin])
```

```python
fechas_iniciales.shape
```

```python
fechas_iniciales = fechas_iniciales.groupby('idaviso').min().reset_index()
```

```python
fechas_iniciales.shape
```

```python
dfPostuMax = dfPostu.loc[:, ['idaviso', 'fechapostulacion']].groupby('idaviso').max().reset_index()
```

```python
dfVMax = dfV.rename(columns={'timestamp':'fechapostulacion', 'idAviso':'idaviso'})\
                        .loc[:, ['idaviso', 'fechapostulacion']].groupby('idaviso').max().reset_index()
```

```python
fechas_finales = pd.concat([dfPostuMax, dfVMax])
```

```python
fechas_finales = fechas_finales.groupby('idaviso').max().reset_index()
```

```python
fechas_finales.shape
```

```python
cantDiasAvisos = fechas_finales
cantDiasAvisos["fechapostulacion"] -= fechas_iniciales["fechapostulacion"]
```

```python
cantDiasAvisos.head(20)
```

```python
def set_hours(row):
    return row.seconds/3600 + row.days*24

cantDiasAvisos["fechapostulacion"] = cantDiasAvisos["fechapostulacion"].apply(set_hours)
```

```python
cantDiasAvisos["cant_horas"] = cantDiasAvisos["fechapostulacion"]
del cantDiasAvisos["fechapostulacion"]
```

```python
dfAvisosDet = pd.merge(dfAvisosDet,cantDiasAvisos,on = "idaviso",how="inner")
```

```python
dfAvisosDet.to_csv("procesado1/avisosDetConFede.csv")
```
