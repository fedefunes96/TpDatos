```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets, linear_model
```

```python
df1 = pd.read_csv("datos_navent_fiuba/fiuba_1_postulantes_educacion.csv")
df2 = pd.read_csv("entrega6/fiuba_1_postulantes_educacion.csv")
dfedu = df1.append(df2)
dfedu.drop_duplicates(inplace = True)
df1 = pd.read_csv("datos_navent_fiuba/fiuba_2_postulantes_genero_y_edad.csv")
df2 = pd.read_csv("entrega6/fiuba_2_postulantes_genero_y_edad.csv")
dfgeneroyedad = df1.append(df2)
dfgeneroyedad.drop_duplicates(inplace = True)
df1 = pd.read_csv("datos_navent_fiuba/fiuba_3_vistas.csv")
df2 = pd.read_csv("entrega6/fiuba_3_vistas.csv")
dfvistas = df1.append(df2)
dfvistas.drop_duplicates(inplace = True)
df1 = pd.read_csv("datos_navent_fiuba/fiuba_4_postulaciones.csv")
df2 = pd.read_csv("entrega6/fiuba_4_postulaciones.csv")
dfpostulaciones = df1.append(df2)
dfpostulaciones.drop_duplicates(inplace = True)
df1 = pd.read_csv("datos_navent_fiuba/fiuba_5_avisos_online.csv")
df2 = pd.read_csv("entrega6/fiuba_5_avisos_online.csv")
dfavisosonline = df1.append(df2)
dfavisosonline.drop_duplicates(inplace = True)
df1 = pd.read_csv("datos_navent_fiuba/fiuba_6_avisos_detalle.csv")
df2 = pd.read_csv("entrega6/fiuba_6_avisos_detalle.csv")
dfavisosdetalle = df1.append(df2)
dfavisosdetalle.drop_duplicates(inplace = True)
```

```python
df1 = pd.merge(dfedu,dfgeneroyedad,on="idpostulante",how="inner")
df2 = pd.merge(dfpostulaciones,dfavisosdetalle,on="idaviso",how="inner")
df = pd.merge(df1,df2,on="idpostulante",how="inner")
```

```python
df["nombre"] = df["nombre"].astype('category').cat.codes
df["estado"] = df["estado"].astype('category').cat.codes
df["sexo"] = df["sexo"].astype('category').cat.codes
df["titulo"] = df["titulo"].astype('category').cat.codes
df["nombre_zona"] = df["nombre_zona"].astype('category').cat.codes
df["ciudad"] = df["ciudad"].astype('category').cat.codes
df["tipo_de_trabajo"] = df["tipo_de_trabajo"].astype('category').cat.codes
df["nivel_laboral"] = df["nivel_laboral"].astype('category').cat.codes
df["nombre_area"] = df["nombre_area"].astype('category').cat.codes
df["denominacion_empresa"] = df["denominacion_empresa"].astype('category').cat.codes
df["sepostulo"] = 1
```

```python
df.head()
```

```python
from sklearn.utils import shuffle
dfpostulaciones["sepostulo"] = 1
dfshuffle = dfpostulaciones[["idaviso","idpostulante"]].set_index("idaviso")
dfshuffle.head()
```

```python
dfshuffle = shuffle(dfshuffle)
dfshuffle.head()
```

```python
dfshuffle.reset_index().set_index(["idaviso","idpostulante"]).head()
```

```python
dfasd = dfpostulaciones[["idaviso","idpostulante"]].set_index(["idaviso","idpostulante"])
dfasd.head()
```

```python
for idx in dfshuffle:
    if (idx in dfasd):
        dfshuffle["sepostulo"] = 1
    else:
        dfshuffle["sepostulo"] = 0
dfshuffle.head()
```

```python
dfentrenamiento = df.sample(frac = 0.70, replace = True)
```

```python
dffeatures = dfentrenamiento[["nombre","estado","sexo","idpais","titulo","nombre_zona","ciudad","tipo_de_trabajo","nivel_laboral","nombre_area","denominacion_empresa"]]
lm = linear_model.LinearRegression(normalize = True)
lm.fit(dffeatures,dfentrenamiento["sepostulo"])
```

```python
dflm = pd.DataFrame(data = dffeatures.columns, columns = ["features"])
dflm["value"] = lm.coef_.round()
dflm
```
