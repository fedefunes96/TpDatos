```python
import pandas as pd
import numpy as npy
import matplotlib as plt
import seaborn as sns
import csv
%matplotlib inline
```

```python
dfAvisosDet = pd.read_csv("fiuba_6_avisos_detalle.csv")

dfPostu = pd.read_csv("fiuba_4_postulaciones.csv")
```

```python
colTemp = dfAvisosDet.filter(items=("nombre_area","idaviso")).groupby("nombre_area").count().sort_values("idaviso",ascending = False)
colTemp = colTemp.loc[colTemp["idaviso"]>20]
colTemp
```

```python
dfTotal = pd.merge(dfAvisosDet,dfPostu,how="right",on="idaviso")
```

```python

otraColTemp = dfTotal.filter(items=("nombre_area","idaviso")).groupby("nombre_area").count().sort_values("idaviso",ascending = False)
otraColTemp
```

```python
colTemp = (otraColTemp/colTemp).sort_values("idaviso",ascending = False).dropna()
colTemp
```

este grafico solo muestra la diferencia entre los con mayor y menor cantidad de postulaciones

```python
x = colTemp.plot(figsize=(10,8))
x.set_title("Cantidad de postulaciones por rubro de trabajo",fontsize=20)
x.set_xlabel('rubro de trabajo',fontsize=15)
x.set_ylabel('Cantidad de postulaciones',fontsize=15)
```

```python
x = colTemp.head(10).plot(kind="bar",figsize=(10,8))
x.set_title("Trabajos con mayor cantidad de postulaciones por rubro de trabajo",fontsize=20)
x.set_xlabel('rubro de trabajo',fontsize=15)
x.set_ylabel('Cantidad de postulaciones',fontsize=15)
```

```python
x = colTemp.tail(10).plot(kind="bar",figsize=(10,8))
x.set_title("Trabajos con menor cantidad de postulaciones por rubro de trabajo",fontsize=20)
x.set_xlabel('rubro de trabajo',fontsize=15)
x.set_ylabel('Cantidad de postulaciones',fontsize=15)
```

```python
colTemp = dfAvisosDet.filter(items=("tipo_de_trabajo","idaviso")).groupby("tipo_de_trabajo").count().sort_values("idaviso",ascending = False)
colTemp = colTemp.loc[colTemp["idaviso"]>20]
colTemp
```

```python

otraColTemp = dfTotal.filter(items=("tipo_de_trabajo","idaviso")).groupby("tipo_de_trabajo").count().sort_values("idaviso",ascending = False)
otraColTemp
```

Los usuarios buscan trabajo por internet, sinembargo ¿no les interesa tato trabajo a distancia? (teletrabajo)

```python
colTemp = (otraColTemp/colTemp).sort_values("idaviso",ascending = False).dropna()
colTemp
```

```python
x = colTemp.plot(kind="bar",figsize=(10,8))
x.set_title("Cantidad de postulaciones por tipo de trabajo",fontsize=20)
x.set_xlabel('tipo_trabajo',fontsize=15)
x.set_ylabel('Cantidad de postulaciones',fontsize=15)
```

```python
colTemp = dfAvisosDet.filter(items=("nivel_laboral","idaviso")).groupby("nivel_laboral").count().sort_values("idaviso",ascending = False)
colTemp = colTemp.loc[colTemp["idaviso"]>20]
colTemp = colTemp.drop("Otro")
colTemp
```
